```
import pandas as pd
import numpy as np
```

```
train_dataset = pd.read_csv('sample_data/california_housing_train.csv')
test_dataset = pd.read_csv('sample_data/california_housing_test.csv')

train = train_dataset.iloc[:, :].values
labels_train = train[:,-1:]
test = test_dataset.iloc[:, :].values
labels_test = test[:, -1:]
```

```
from sklearn.preprocessing import StandardScaler
```

```
sc = StandardScaler()
sc = sc.fit(train[:, :-1])
train = sc.transform(train[:, :-1])
test = sc.transform(test[:,:-1])
```

```
train.shape
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import TensorBoard
from tensorboardcolab import TensorBoardColab, TensorBoardColabCallback
```

```
model = Sequential()
model.add(Dense(units=4,input_dim=8, activation='relu', kernel_initializer='uniform'))
model.add(Dense(units=1))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```
tbc=TensorBoardColab()

model.fit(train,labels_train,
          validation_data=(test,labels_test),
          epochs=100, batch_size=100,
          callbacks=[TensorBoardColabCallback(tbc)])
```
